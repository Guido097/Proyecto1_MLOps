```python
import pandas as pd 
import ast
```

```python
#Ya que el archivo json no se encuentra bien formateado vamos a leer linea por linea
# para transformarlo en un dataframe con el que podamos trabajar.
rows = []
with open('australian_users_items.json',encoding='MacRoman') as f: 
  for line in f.readlines():
    rows.append(ast.literal_eval(line)) 

df_items = pd.DataFrame(rows)
```

```python
df_items.head()
```

```python
df_items.shape
```

```python
#Guardamos para no tener que leer el .json de vuelta ante un error.
df_items.to_csv('items.csv', index=False)
```

```python
#Checkeamos el promedio para saber con que cantidad de elementos vamos a trabajar.
promedio = df_items["items_count"].mean()
print("El promedio es:", promedio)
```

```python
# Define una función para extraer los 'id' de cada fila
def extract_ids(row):
    if isinstance(row, list):
        return [item['item_id'] for item in row]
    else:
        return []

# Aplica la función a cada fila y crea una nueva columna llamada 'ids'
df_items['ids'] = df_items['items'].apply(extract_ids)

# Ahora, 'df_items' contiene una nueva columna llamada 'ids' con los valores 'id' para cada fila
df_items.head(10)
```

```python
#Vamos a aplicar lo mismo para extraer las horas jugadas por cada app.
def extract_ptf(row):
    if isinstance(row, list):
        return [item['playtime_forever'] for item in row]
    else:
        return []

df_items['playtimeforever'] = df_items['items'].apply(extract_ptf)

df_items.head()
```

```python
#Nos quedamos solo con algunas columnas para trabajar mejor.
columns = ['user_id', 'items_count', 'steam_id','ids','playtimeforever']
df_items = df_items[columns]
```

```python
df_items.head()
```

```python
#Desanidamos las columnas 'ids' y 'playtimeforever'
df_items = df_items.explode(['ids', 'playtimeforever'])
```

```python
df_items.head()
```

```python
#Vamos a quedarnos solo con las columnas que nos hacen falta.
columns = ['user_id','steam_id','ids','playtimeforever']
df_items = df_items[columns]
df_items.head()
```

```python
df_items.shape
```

```python
#Para reducir los 5M de registros saco todos los que no tengan horas jugadas.
df_items = df_items[df_items['playtimeforever'] != 0]
```

```python
df_items.head()
```

```python
df_items.shape
```

```python
#Para seguir reduciendo la dimension del dataframe, vamos a sacar tambien
#a los que tengan 2 o menos horas jugadas, ya que por politicas de la plataforma
# 'Steam' esas personas podrian devolver los juegos.
df_items = df_items[df_items['playtimeforever'] > 2]
```

```python
df_items.shape
```

```python
#Guardamos el dataframe que vamos a seguir utilizando.
df_items.to_csv('users_ptf.csv', index=False)
```

```python
#Me voy a quedar con solo 2 columnas para trabajar el tiempo de juego general.
df_time = df_items[['ids', 'playtimeforever']]
```

```python
#Realizo un 'groupby' para que me quede el total de horas por 'id'.
df_time = df_time.groupby('ids', as_index=False)['playtimeforever'].sum()
```

```python
df_time.head()
```

```python
#Guardamos el dataframe que vamos a seguir utilizando.
df_time.to_csv('id_time.csv', index=False)
```

```python
df_names = pd.read_csv('id_names.csv')
df_names.head()
```

```python
#Renombramos la columna para que se facilite el merge.
df_items.rename(columns={'ids': 'id'}, inplace=True)
```

```python
#Convertimos en integer para que permita el merge.
df_names['id'] = df_names['id'].astype(int)
```

```python
#Convertimos en integer para que permita el merge.
df_items['id'] = df_items['id'].astype(int)
```

```python
#Hacemos un merge para agregar los nombres de las apps y facilitar 
# las funciones de los endpoints.
df_items = pd.merge(df_items, df_names, on='id', how='inner')
```

```python
df_items.head()
```

```python
df_genre = pd.read_csv('id_genre.csv')
df_genre.head()
```

```python
df_items = pd.merge(df_items, df_genre, on='id', how='inner')
```

```python
#Guardamos el archivo final.
df_items.to_csv('items.csv', index= False)
```
