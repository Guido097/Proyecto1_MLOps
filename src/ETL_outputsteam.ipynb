```python
import pandas as pd
import numpy as np
import json
from dateutil.parser import parse
```

```python
#Leemos el archivo '.json'
df_games = pd.read_json('output_steam_games.json', lines=True)
```

```python
#Dropeamos los nulos de la columna 'id'
df_games = df_games.dropna(subset='id')
```

```python
df_games.head()
```

```python
#Checkeamos la dimension del dataframe
df_games.shape
```

```python
#Checkeamos el formato de cada columna del dataframe.
df_games.dtypes
```

```python
#Dropeamos las columnas que no nos interesan.
df_games = df_games.drop(['user_id','steam_id','items','items_count'], axis=1)
```

```python
#Definimos el formato de la columna generos como string.
df_games['genres'] = df_games['genres'].astype(str)
```

```python
df_games.head(10)
```

```python
#Formateamos, extraemos y generamos valores estilo booleanos para los generos.
df_games['genres'] = df_games['genres'].str.replace(r'\[|\]', '', regex=True)
df_games['genres'] = df_games['genres'].str.replace(']','',regex=True).replace("'","",regex=True)
df_games['genres'] = df_games['genres'].str.replace(';',',',regex=True)
genres_df = df_games['genres'].str.get_dummies(', ')
genres_df.columns = [col.capitalize() for col in genres_df.columns]
df_games = pd.concat([df_games, genres_df], axis=1)
```

```python
df_games.head(10)
```

```python
#Formateamos la columna 'release_date' y extraemos el anio de lanzamiento.
df_games["release_date"] = pd.to_datetime(df_games["release_date"], errors='coerce')
df_games["release_date"] = df_games["release_date"].apply(lambda x: parse(str(x)).strftime('%Y-%m-%d') if pd.notnull(x) else None)
df_games["release_date"] = pd.to_datetime(df_games["release_date"], errors='coerce')
df_games["release_year"] = df_games["release_date"].dt.year
```

```python
df_games.head(10)
```

```python
#Imprimimos los valores unicos para ver los anios de lanzamiento.
unique_release_years = df_games['release_year'].unique()
unique_release_years = sorted(unique_release_years)
print(unique_release_years)
```

```python
#Como vimos que tenemos fechas nulas, las dropeamos para alivianar el dataframe ya que no nos serviran esos registros.
df_games.dropna(subset=["release_date"], inplace=True)
df_games.drop_duplicates(subset=['id'])
```

```python
#Reemplazamos los precios como "free to play" por 0 y los "Starting at $449.00" por el valor correspondiente.
df_games['price'] = df_games['price'].replace(['Free To Play','Install Now','Third-party','Install Theme','Play Now','Free Movie','Free to Try', 'Play WARMACHINE: Tactics Demo','Play the Demo', 'Free','Free to Use', 'Free HITMAN™ Holiday Pack','Free to Play','Play for Free!','Free Mod','Free Demo', 'nan'], '0')
df_games['price'] = df_games['price'].replace(['Starting at $449.00'], 449.00)
df_games['price'] = df_games['price'].replace(['Starting at $499.00'], 499.00)
```

```python
#Imprimimos los valores unicos de la columna 'price'
unique_prices = df_games['price'].unique()
print(unique_prices)
```

```python
#Formateamos la columna 'price'
df_games['price'] = pd.to_numeric(df_games['price'], errors='coerce')
```

```python
#Rellenamos los nulos con '0'
df_games['price'].fillna(0, inplace=True)
```

```python
df_games.head()
```

```python
#Rellenamos los nulos con '0'
df_games['discount_price'].fillna(0, inplace=True)
```

```python
df_games.head()
```

```python
#Dropeamos todas las columnas que no nos hacen falta
df_games.drop('title', axis=1, inplace=True)
df_games.drop('price', axis=1, inplace=True)
df_games.drop('early_access', axis=1, inplace=True)
df_games.drop('genres', axis=1, inplace=True)
df_games.drop('discount_price', axis=1, inplace=True)
df_games.drop('url', axis=1, inplace=True)
df_games.drop('reviews_url', axis=1, inplace=True)
df_games.drop('release_date', axis=1, inplace=True)
df_games.drop('publisher', axis=1, inplace=True)

```

```python
df_games.head()
```

```python
#Ordenamos el dataframe en base a los 'id' de cada juego.
df_games = df_games.sort_values(by='id')
```

```python
#Y como ordenamos nuevamente el dataframe reiniciamos el indice
# a la vez que eliminamos el indice anterior
df_games.reset_index(inplace=True)
df_games.drop('index', axis=1, inplace=True)

```

```python
df_games.head()
```

```python
#Imprimimos valores unicos de la columna 'metascore'
metascore = df_games['metascore'].unique()
print(metascore)
```

```python
#Limpiamos los valores que no nos sirven.
df_games['metascore'] = df_games['metascore'].replace(['None','NA'], 0)
```

```python
#Limpiamos los valores que no nos sirven.
df_games['metascore'].fillna(0, inplace=True)
```
